```python
import pandas as pd
from tqdm import tqdm
import os
import xml.etree.ElementTree as ET
```

```python
# where the guesses are
# C:\Users\piano\Desktop\PhD Rotation--Nick\data\Deepcadrme_guesses

# these are the deepcadrme outputs (in a .xml file format)
# it will look like DRUGNAME.xml

filepath = '../data/deepcadrme/guess_xml/'
filenames = os.listdir(filepath)
```

```python
"./Mentions/Mention[@type='AdverseReaction']" and "./Mentions/Mention[@section='S1']"
```

```python
current_file = []
term_extracted = []
start_intext = []
len_intext = []
sec_extracted = []
type_extracted = []

for file in tqdm(filenames):
    tree = ET.parse(filepath+ file)
    root = tree.getroot()
    
    # grab text from label--don't need to do this yet
    # label_text = root.findall("./Text/Section[@name='adverse reactions']")[0].text
    
    section_map = dict()
    
    for section in root.findall("./Text/Section"):
        #print(section.attrib['name'], section.attrib['id'])
        section_map[section.attrib['id']] = section.attrib['name']
    
    for adr_mention in root.findall("./Mentions/Mention"):
        current_file.append(file)
        sec_extracted.append(section_map[adr_mention.attrib['section']])
        type_extracted.append(adr_mention.attrib['type'])
        term_extracted.append(adr_mention.attrib['str']) # this will print the whole thing
        start_intext.append(adr_mention.attrib['start'])
        len_intext.append(adr_mention.attrib['len']) # this will just print the value for the 'start' attrib
```

```python
df = pd.DataFrame(zip(current_file, sec_extracted, type_extracted, term_extracted, start_intext, len_intext), columns = ["file", "section", "type", "term", "start", "len"])
df.to_csv("../data/Deepcadrme_guess_terms.csv")
```
