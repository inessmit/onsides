### Tables of text from FDA drug labels
From 45,110 drug labels, we extract 21,939 with sections related to "use in specific populations". Extract "SP" / "AR" / "WP" / "BW" sections.

```python
#extract SP section
sp_pop_list = []
for n in range(5):
  f = glob(folder+'spl/rx/dm_spl_release_human_rx_part{}/prescription/*'.format(str(n+1)))
  for i in tqdm(f):
    with open(i, 'r') as fi:
      j = ujson.load(fi)
    if 'SP' in j['sections']:
      sp_pop_list.append([j['set_id'], j['label_id'], j['sections']['SP']])
sp_pop_df = pd.DataFrame(sp_pop_list, columns = ['set_id', 'label_id', 'sp_text'])
sp_pop_df.to_csv(folder+'special_populations/sp_section_text_all.csv', index = False)
sp_pop_df.head()
```

### Map Terms

```python
#read in meddra database
externals = '/content/drive/MyDrive/pop_pharmacogenomics/onsides_intl/external_data/'
umls_map = pd.read_csv(externals+'umls_meddra_en.csv')
umls_map['STR'] = umls_map.STR.apply(lambda x: x.lower())
umls_map = umls_map[['STR', 'TTY', 'SDUI']].drop_duplicates(subset='STR')
print(umls_map.shape[0])
umls_map['len'] = umls_map['STR'].apply(lambda x: len(x))
umls_map = umls_map[(umls_map.TTY.isin(['PT', 'LLT']))&(umls_map['len'] >= 5)]
print(umls_map.shape[0])
umls_map_dict = dict(zip(umls_map.STR, umls_map.SDUI))
umls_map.head(1)
```

#### Boxed Warnings

```python
bw_pop_df = pd.read_csv(folder+'special_populations/bw_section_ped_text_filtered.csv',)
bw_pop_df.head(1)
```

```python
exact_terms = []
for i, row in tqdm(bw_pop_df.iterrows()):
  label_id = row['label_id']
  text = row['section_text']
  found_terms = list()
  for mdr_term in umls_map_dict.keys():
    if text.find(mdr_term) == -1:
      continue
    else:
      li = text.split(mdr_term)
      start_pos = 0
      for i in range(len(li)-1):
          # the occurrence of the word is at the end of the previous string
          start_pos = sum([len(li[j]) for j in range(i+1)]) + i*len(mdr_term)
          if not mdr_term == text[start_pos:(start_pos+len(mdr_term))]:
              raise Exception(f" mdr_term: '{mdr_term}', term_in_text: '{text[start_pos:(start_pos+len(mdr_term))]}'")
          found_terms.append((mdr_term, umls_map_dict[mdr_term], start_pos, len(mdr_term)))
  exact_terms.append([label_id, found_terms, text])
```

```python
exact_terms_df = pd.DataFrame(exact_terms, columns=['label_id', 'found_terms', 'pop_text'])
exact_terms_df = exact_terms_df.explode('found_terms')
exact_terms_df['found_term'] = exact_terms_df['found_terms'].apply(lambda x: x[0] if str(x) != 'nan' else None)
exact_terms_df['meddra_id'] = exact_terms_df['found_terms'].apply(lambda x: x[1] if str(x) != 'nan' else None)
exact_terms_df['location'] = exact_terms_df['found_terms'].apply(lambda x: x[2] if str(x) != 'nan' else None)
exact_terms_df = exact_terms_df[exact_terms_df.found_term.notna()].drop(['found_terms'], axis = 1)
exact_terms_df.to_csv(folder+'special_populations/bw_section_ped_text_exact_terms.csv', index=False)
exact_terms_df.head()
```

```python
def parse_pop(input_list):
    result = []
    current_title = None
    current_content = []

    for item in input_list:
        if item.startswith('8.'):
            # If the item is a title, append the previous title and its associated content to the result
            if current_title is not None:
                result.append([current_title, current_content])
            # Update the current title and reset the current content list
            current_title = item
            current_content = []
        else:
            # If the item is not a title, append it to the current content list
            current_content.append(item)

    # Append the last title and its associated content to the result
    if current_title is not None:
        result.append([current_title, current_content])

    return result

sp_pop_df['pop_list'] = sp_pop_df.sp_text.apply(lambda x: parse_pop([i.strip() for i in x.split('\n') if i.strip() != '']))
sp_pop_df = sp_pop_df[['set_id', 'label_id', 'pop_list']].explode('pop_list')
sp_pop_df['pop'] = sp_pop_df.pop_list.apply(lambda x: x[0] if str(x) != 'nan' else None)
sp_pop_df['pop_text'] = sp_pop_df.pop_list.apply(lambda x: x[1] if str(x) != 'nan' else None)
sp_pop_df = sp_pop_df[sp_pop_df['pop'].notna()]
```

```python
#read in meddra database
externals = 'external_data/'
umls_map = pd.read_csv(externals+'umls_meddra_en.csv')
umls_map['STR'] = umls_map.STR.apply(lambda x: x.lower())
umls_map = umls_map[['STR', 'TTY', 'SDUI']].drop_duplicates(subset='STR')
umls_map_dict = dict(zip(umls_map.STR, umls_map.SDUI))
umls_map.head(1)
```

```python
exact_terms = []
for i, row in tqdm(ped_use.iterrows()):
  label_id = row['label_id']
  title = row['pop']
  text = ' '.join(row['pop_text']).lower()
  found_terms = list()
  for mdr_term in umls_map_dict.keys():
    if text.find(mdr_term) == -1:
      continue
    else:
      li = text.split(mdr_term)
      start_pos = 0
      for i in range(len(li)-1):
          # the occurrence of the word is at the end of the previous string
          start_pos = sum([len(li[j]) for j in range(i+1)]) + i*len(mdr_term)
          if not mdr_term == text[start_pos:(start_pos+len(mdr_term))]:
              raise Exception(f" mdr_term: '{mdr_term}', term_in_text: '{text[start_pos:(start_pos+len(mdr_term))]}'")
          found_terms.append((mdr_term, umls_map_dict[mdr_term], start_pos, len(mdr_term)))
  exact_terms.append([label_id, found_terms, text])
```

```python
exact_terms_df = pd.DataFrame(exact_terms, columns=['label_id', 'found_terms', 'pop_text'])
exact_terms_df = exact_terms_df.explode('found_terms')
exact_terms_df['len'] = exact_terms_df['found_terms'].apply(lambda x: x[3] if str(x) != 'nan' else None)
exact_terms_df = exact_terms_df[exact_terms_df['len'] >= 5]
exact_terms_df['found_term'] = exact_terms_df['found_terms'].apply(lambda x: x[0] if str(x) != 'nan' else None)
exact_terms_df['meddra_id'] = exact_terms_df['found_terms'].apply(lambda x: x[1] if str(x) != 'nan' else None)
exact_terms_df['location'] = exact_terms_df['found_terms'].apply(lambda x: x[2] if str(x) != 'nan' else None)
exact_terms_df = exact_terms_df[exact_terms_df.found_term.notna()].drop(['found_terms', 'len'], axis = 1)
exact_terms_df.head()
```

```python
building_strings = []
for i, row in tqdm(exact_terms_df.iterrows()):
  term, label_id, start_pos, ar_text = row['found_term'], row['label_id'], int(row['location']), row['pop_text']
  #default settings
  nwords, prop_before = 125, 0.125
  #pull the full text
  #ar_text = ped_use_dict[label_id]

  term_nwords = len(term.split())
  size_before = max(int((nwords-2*term_nwords)*prop_before), 1)
  size_after = max(int((nwords-2*term_nwords)*(1-prop_before)), 1)

  before_text = ar_text[:start_pos]
  after_text = ar_text[(start_pos+term_nwords):]

  before_parts = before_text.split()[-1*size_before:]
  after_parts = after_text.split()[:size_after]

  li = [term]
  li.extend(before_parts)
  li.append('EVENT')
  li.extend(after_parts)
  example_string = ' '.join(li)
  building_strings.append(example_string)
exact_terms_df['string'] = building_strings
exact_terms_df.head() #3328739
```
